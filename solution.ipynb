Columns with missing values: director, cast, country, date_added, rating, duration

```python
df.isnull().sum()
```

```python
# Replace missing values in director column
df['director'].fillna(value='Unknown', inplace= True) 
```

```python
# Replace missing values in cast column
df['cast'].fillna(value='Unknown', inplace= True)
```

```python
# Replace missing values in country column
#df['cast'].fillna(value='Unknown', inplace= True)
df.fillna({'country': 'Unknown'}, inplace= True)
```

```python
#drop missing values in date_added, rating and duration columns
df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True) 
```
